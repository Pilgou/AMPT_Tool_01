```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Exemple série temporelle
date_rng = pd.date_range(start="2025-08-28", periods=100, freq="min")
data = np.sin(np.linspace(0, 20, 100)) + np.random.normal(0, 0.2, 100)
data[20] = 5   # anomalie volontaire
data[70] = -4  # anomalie volontaire

df = pd.DataFrame({"time": date_rng, "value": data})
#Creation erreur au niveau de df
df.loc[70, "value"] = -1
```

```python
    # Tracé data
df["value"].plot(figsize=(8,4), title="data")
plt.xlabel("time")
plt.ylabel("Data")
plt.legend()
#plt.grid(True)
plt.show()
```

```python
# Détection par Z-score
mean = df["value"].mean()
std = df["value"].std()
df["z_score"] = (df["value"] - mean) / std
df["anomaly"] = df["z_score"].apply(lambda x: abs(x) > 3)

print(df[df["anomaly"]])
```

```python
    # Tracé Z Score
df["z_score"].plot(figsize=(8,4), title="z_score")
plt.xlabel("time")
plt.ylabel("z_score")
plt.legend()
#plt.grid(True)
plt.show()
```
