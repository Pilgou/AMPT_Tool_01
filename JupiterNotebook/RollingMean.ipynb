```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Exemple s√©rie temporelle
date_rng = pd.date_range(start="2025-08-28", periods=100, freq="min")
data = np.sin(np.linspace(0, 20, 100)) + np.random.normal(0, 0.2, 100)
data[20] = 2   # anomalie volontaire
data[70] = -4  # anomalie volontaire

df = pd.DataFrame({"time": date_rng, "value": data})
#Creation erreur au niveau de df
df.loc[71, "value"] = -1
```

```python
    # Trac√© data
df["value"].plot(figsize=(8,4), title="data")
plt.xlabel("time")
plt.ylabel("Data")
plt.legend()
#plt.grid(True)
plt.show()
```

```python
#window = 5
df["rolling_mean"] = df["value"].rolling(window = 5).mean()
df["rolling_std"] = df["value"].rolling(window = 5).std()

# Anomalie si √©cart > 3 * √©cart-type local
df["anomaly"] = abs(df["value"] - df["rolling_mean"]) > 3 * df["rolling_std"]
```
